```python
troca_evolucao = {1:'Curados', 2:'Óbitos', 9:'Ignorado'}
```

```python
df.loc[(df.EVOLUCAO.isnull()), 'EVOLUCAO'] = 9
df['EVOLUCAO'] = df['EVOLUCAO'].apply(lambda x: troca_evolucao[x])
```

```python
df.sample(10)
```

```python
df.query('CLASSI_FIN == 5 ').groupby('EVOLUCAO').size().sort_values(ascending=False).plot(kind='bar', color='RED',title = 'Evolução dos casos COVID-19').set_ylabel('Casos',fontsize=14)
```
